```python
missing_values = ((df_merged.isna().sum() / df_merged.shape[0]) * 100).sort_values()
print('Процент пропущенных значений:', missing_values)
```

```python
df_clean = df_merged.copy() # Удалим колонки с числом пропущенных значений более 25%
df_clean = df_clean.drop(columns=df_merged.loc[:, missing_values > 25])
```

```python
def boundaries(df_col):
    q25 = df_col.quantile(0.25)
    q75 = df_col.quantile(0.75)
    iqr = q75 - q25
    return q75 + 1.5 * iqr, q25 - 1.5 * iqr
```

```python
cols = ['loser_age', 'loser_rank', 'loser_rank_points', 'match_num',  # Колонки, в которых будем искать выбросы
        'winner_age', 'winner_rank', 'winner_rank_points']  
df_bound = {i: boundaries(df_clean[i]) for i in cols if df_clean[i].dtype != 'object'} # получим в числовой форме границы усов для графика boxplot
df_bound
```
